```python



import torch
import torch.nn.functional as F
from torch.optim import Adam
import scipy.sparse as sp
import numpy as np
import os
os.chdir("/home/mmip/Documents/code/python/simulation")
import time
import matplotlib.pyplot as plt
from tqdm import tqdm
import matplotlib.patches as mpatches
from preprocessing import *
from model import *
from HSIC import *
import networkx as nx
import scipy

#os.environ['CUDA_VISIBLE_DEVICES'] = ""

import pandas
       
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

import networkx as nx
```

```python
n01=83
n02=306
z1x_index = np.random.binomial(1,0.5,n01).reshape(-1,1)
z1x=  np.array([-2,2])[z1x_index]
z1x = np.random.normal(loc=z1x)
z1y = np.random.normal(size=(n01,1))
z1 = np.concatenate([z1x,z1y],axis=1)

z2x_index = np.random.binomial(1,0.5,n02).reshape(-1,1)
z2x=  np.array([-2.5,2.5])[z2x_index]
z2x = np.random.normal(loc=z2x)
z2y=  np.array([0,2])[z2x_index]
z2y = np.random.normal(loc=z2y,size = (n02,1))
z2=np.concatenate([z2x,z2y],axis=1)

score0=distance_decode(torch.Tensor(z1),torch.Tensor(z2))
bipartite_net=torch.bernoulli(score0)  
adj_plot = sp.csr_matrix(bipartite_net) 
G=nx.algorithms.bipartite.from_biadjacency_matrix(adj_plot)
position = {k: np.vstack([z1,z2])[k] for k in G.nodes.keys()}

fig, ax = plt.subplots()
nx.draw_networkx(G,
                 position,
                 node_size=4,
                 with_labels=False,
                 node_color = n01*["#1f77b4"]+n02*["red"],
                 edge_color = (0.75,0.75,0.75),
                 ax=ax)
plt.title("Simulated latent space")
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)




z1z = np.random.normal(0,size=(n01,1))*1.5
z2z = np.random.normal(0,1,size=(n02,1))*1
#z1z[z1x_index>0] = z1z[z1x_index>0]+0.75
#z1z[z1x_index==1] = z1z[z1x_index==1]*2
#z1z=np.random.normal(np.array([-0.5,0.5])[z1x_index])
#z2z=np.random.normal(np.array([0.5,-0.5])[z2x_index])

z1 = np.concatenate([z1x,z1y,z1z],axis=1)
z2 =np.concatenate([z2x,z2y,z2z],axis=1)


score_obs=distance_decode(torch.Tensor(z1),torch.Tensor(z2))

net0=torch.bernoulli(score_obs)
adj = sp.csr_matrix(net0) 

adj0 = net0


features01 = np.eye(adj0.shape[0])
features02 = np.eye(adj0.shape[1])
features1 = sp.csr_matrix(features01) 
features2 = sp.csr_matrix(features02) 

    
features1 = sparse_to_tuple(features1.tocoo())
features2 = sparse_to_tuple(features2.tocoo())
adj_train, train_edges, val_edges, val_edges_false, test_edges, test_edges_false = mask_test_edges(adj0)
adj_norm = preprocess_graph(adj_train)

n=adj.shape[0]
    # Create Model
pos_weight = float(adj.shape[0] * adj.shape[1] - adj.sum()) / adj.sum()
norm = adj.shape[0] * adj.shape[1] / float((adj.shape[0] * adj.shape[1] - adj.sum()) * 2)


adj_label = adj_train 
adj_label = sparse_to_tuple(adj_label)



adj_norm = torch.sparse.FloatTensor(torch.LongTensor(adj_norm[0].T), 
                            torch.FloatTensor(adj_norm[1]), 
                            torch.Size(adj_norm[2]))
adj_label = torch.sparse.FloatTensor(torch.LongTensor(adj_label[0].T), 
                            torch.FloatTensor(adj_label[1]), 
                            torch.Size(adj_label[2]))


features1 = torch.sparse.FloatTensor(torch.LongTensor(features1[0].T), 
                            torch.FloatTensor(features1[1]), 
                            torch.Size(features1[2]))
features2 = torch.sparse.FloatTensor(torch.LongTensor(features2[0].T), 
                            torch.FloatTensor(features2[1]), 
                            torch.Size(features2[2]))


weight_mask = adj_label.to_dense().view(-1) == 1
weight_tensor = torch.ones(weight_mask.size(0)) 
weight_tensor[weight_mask] = pos_weight


G=nx.algorithms.bipartite.from_biadjacency_matrix(adj)
position = {k: np.vstack([z1[:,[0,2]],z2[:,[0,2]]])[k] for k in G.nodes.keys()}
fig, ax = plt.subplots()
nx.draw_networkx(G,
                 position,
                 node_size=4,
                 with_labels=False,
                 node_color = n01*["#1f77b4"]+n02*["red"],
                 edge_color = (0.75,0.75,0.75),
                 ax=ax)
plt.title("Simulated latent space")
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)

G=nx.algorithms.bipartite.from_biadjacency_matrix(adj)
position = {k: np.vstack([z1[:,[1,2]],z2[:,[1,2]]])[k] for k in G.nodes.keys()}
fig, ax = plt.subplots()
nx.draw_networkx(G,
                 position,
                 node_size=4,
                 with_labels=False,
                 node_color = n01*["#1f77b4"]+n02*["red"],
                 edge_color = (0.75,0.75,0.75),
                 ax=ax)
plt.title("Simulated latent space")
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
    
    
    #torch.manual_seed(2)
model = VBGAE(adj_norm,83,306)
init_parameters(model)

optimizer = Adam(model.parameters(), lr=args.learning_rate)

# train model
roclist = []
loss_list= []
    
    
    #torch.manual_seed(4)
pbar = tqdm(range(1000),desc = "Training Epochs")
for epoch in pbar:
    t = time.time()

    A_pred,Z1,Z2 = model(features1,features2)
    optimizer.zero_grad()
    loss = norm*F.binary_cross_entropy(A_pred.view(-1), adj_label.to_dense().view(-1), weight = weight_tensor)

    kl_divergence = 0.5/ A_pred.size(0) *( (1 + 2*model.logstd1 - model.mean1**2 - torch.exp(model.logstd1)**2).sum(1).mean()+
                                          (1 + 2*model.logstd2 - model.mean2**2 - torch.exp(model.logstd2)**2).sum(1).mean())
    loss -= kl_divergence
    loss.backward()
    optimizer.step()
    

    val_roc, val_ap = get_scores(val_edges, val_edges_false, A_pred)
    
    roclist.append(val_roc)
    loss_list.append(loss.item())

    pbar.set_postfix({"train_loss=": "{:.5f}".format(loss.item()),
                      'val_roc=': val_roc})


A_pred,Z1,Z2 = model(features1,features2)

AUC1, test_ap = get_scores(test_edges, test_edges_false, A_pred)
print("1) End of training!", "test_roc=", "{:.5f}".format(AUC1),
      "test_ap=", "{:.5f}".format(test_ap))

    
    #torch.manual_seed(7)#16
model2 = VBGAE(adj_norm,83,306)
init_parameters(model2)
    #S = torch.Tensor(features01[:,Z]).reshape(-1,1)
S1 = torch.Tensor(z1z).reshape(-1,1)
S1 = (S1-S1.mean())/S1.std()
S2 = torch.Tensor(z2z).reshape(-1,1)

    
optimizer = Adam(model2.parameters(), lr=args.learning_rate)
    
    # train model
roclist = []
loss_list= []
indep_list = []

    #torch.manual_seed(2)#3
pbar = tqdm(range(1000),desc = "Training Epochs")
for epoch in pbar:
    t = time.time()

    A_pred,Z1,Z2 = model2(features1,features2)
    optimizer.zero_grad()
    loss = norm*F.binary_cross_entropy(A_pred.view(-1), adj_label.to_dense().view(-1), weight = weight_tensor)

    kl_divergence = 0.5/ A_pred.size(0) *( (1 + 2*model2.logstd1 - model2.mean1**2 - torch.exp(model2.logstd1)**2).sum(1).mean()+
                                          (1 + 2*model2.logstd2 - model2.mean2**2 - torch.exp(model2.logstd2)**2).sum(1).mean())
    loss -= kl_divergence
    independance =2*(RFF_HSIC(model2.mean1,S1)) #+(RFF_HSIC(model2.mean2,S2)) #0.02
    #independance =2*torch.log(RFF_HSIC(Z1,S))#0.02

    loss += 1*independance
    loss.backward()
    optimizer.step()
    

    val_roc, val_ap = get_scores(val_edges, val_edges_false, A_pred)
    
    roclist.append(val_roc)
    loss_list.append(loss.item())
    indep_list.append(independance.item())

    pbar.set_postfix({"train_loss=": "{:.5f}".format(loss.item()),
                      'val_roc=': val_roc,'log_HSIC =': independance.item()})


AUC2, test_ap = get_scores(test_edges, test_edges_false, A_pred)
print("1) End of training!", "test_roc=", "{:.5f}".format(AUC2),
      "test_ap=", "{:.5f}".format(test_ap))



stat1 = HSIC_stat(model2.mean1,S1)
x = np.linspace(0, 0.4, 100)
y = stats.gamma.cdf(x,stat1[3].item(),scale=stat1[4].item())
plt.figure()
plt.plot(x,y)
plt.axvline(x = stat1[0].item()*n)
print(stats.gamma.sf(stat1[0].item()*n, stat1[3].item(), scale=stat1[4].item()))

stat1 = HSIC_stat(model2.mean2,S2)
x = np.linspace(0, 0.4, 100)
y = stats.gamma.cdf(x,stat1[3].item(),scale=stat1[4].item())
plt.figure()
plt.plot(x,y)
plt.axvline(x = stat1[0].item()*n)
print(stats.gamma.sf(stat1[0].item()*n, stat1[3].item(), scale=stat1[4].item()))

A_pred,Z1,Z2 = model(features1,features2)
A_pred_fair,Z1,Z2 = model2(features1,features2)  


print(scipy.stats.spearmanr(score0.detach().numpy().reshape(-1,1),A_pred.detach().numpy().reshape(-1,1)))
print(scipy.stats.spearmanr(score0.detach().numpy().reshape(-1,1),A_pred_fair.detach().numpy().reshape(-1,1)))

A_pred3 = distance_decode(model.mean1,model.mean2).detach().numpy()
A_pred_fair3 = distance_decode(model2.mean1,model2.mean2).detach().numpy()

print(scipy.stats.spearmanr(score0.detach().numpy().reshape(-1,1),A_pred3.reshape(-1,1)))
print(scipy.stats.spearmanr(score0.detach().numpy().reshape(-1,1),A_pred_fair3.reshape(-1,1)))

```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Apr 11 15:32:41 2024

@author: mmip
"""


RES = pandas.DataFrame(columns=["AUC1","AUC2",
                                "pHSIC1","pHSIC2","SP01","SP02",
                                "SP1","SP2"],index=range(50))

np.random.seed(1)
for k in range(50): 
    n01=83
    n02=306
    z1x_index = np.random.binomial(1,0.5,n01).reshape(-1,1)
    z1x=  np.array([-2,2])[z1x_index]
    z1x = np.random.normal(loc=z1x)
    z1y = np.random.normal(size=(n01,1))
    z1 = np.concatenate([z1x,z1y],axis=1)
    
    z2x_index = np.random.binomial(1,0.5,n02).reshape(-1,1)
    z2x=  np.array([-2.5,2.5])[z2x_index]
    z2x = np.random.normal(loc=z2x)
    z2y=  np.array([0,2])[z2x_index]
    z2y = np.random.normal(loc=z2y,size = (n02,1))
    z2=np.concatenate([z2x,z2y],axis=1)
    
    score0=distance_decode(torch.Tensor(z1),torch.Tensor(z2))
    bipartite_net=torch.bernoulli(score0)  
    
    z1z = np.random.normal(0,size=(n01,1))*1.5
    z2z = np.random.normal(0,size=(n02,1))
    #z1z[z1x_index>0] = 1*z1z[z1x_index>0]
    #z1z=np.random.normal(np.array([-0.5,0.5])[z1x_index])*1.5
    #z2z=np.random.normal(np.array([0.5,-0.5])[z2x_index])
    
    z1 = np.concatenate([z1x,z1y,z1z],axis=1)
    z2 =np.concatenate([z2x,z2y,z2z],axis=1)
    score_obs=distance_decode(torch.Tensor(z1),torch.Tensor(z2))
    
    net0=torch.bernoulli(score_obs)
    adj = sp.csr_matrix(net0) 
    
    adj0 = net0

    
    features01 = np.eye(adj0.shape[0])
    features02 = np.eye(adj0.shape[1])
    features1 = sp.csr_matrix(features01) 
    features2 = sp.csr_matrix(features02) 
    
    
    features1 = sparse_to_tuple(features1.tocoo())
    features2 = sparse_to_tuple(features2.tocoo())
    adj_train, train_edges, val_edges, val_edges_false, test_edges, test_edges_false = mask_test_edges(adj0)
    adj_norm = preprocess_graph(adj_train)
    
    n=adj.shape[0]
    # Create Model
    pos_weight = float(adj.shape[0] * adj.shape[1] - adj.sum()) / adj.sum()
    norm = adj.shape[0] * adj.shape[1] / float((adj.shape[0] * adj.shape[1] - adj.sum()) * 2)
    
    
    adj_label = adj_train 
    adj_label = sparse_to_tuple(adj_label)
    
    
    
    adj_norm = torch.sparse.FloatTensor(torch.LongTensor(adj_norm[0].T), 
                                torch.FloatTensor(adj_norm[1]), 
                                torch.Size(adj_norm[2]))
    adj_label = torch.sparse.FloatTensor(torch.LongTensor(adj_label[0].T), 
                                torch.FloatTensor(adj_label[1]), 
                                torch.Size(adj_label[2]))
    
    
    features1 = torch.sparse.FloatTensor(torch.LongTensor(features1[0].T), 
                                torch.FloatTensor(features1[1]), 
                                torch.Size(features1[2]))
    features2 = torch.sparse.FloatTensor(torch.LongTensor(features2[0].T), 
                                torch.FloatTensor(features2[1]), 
                                torch.Size(features2[2]))
    
    
    weight_mask = adj_label.to_dense().view(-1) == 1
    weight_tensor = torch.ones(weight_mask.size(0)) 
    weight_tensor[weight_mask] = pos_weight
    
    
    
    
    
    #torch.manual_seed(2)
    model = VBGAE(adj_norm,83,306)
    init_parameters(model)
    
    optimizer = Adam(model.parameters(), lr=args.learning_rate)
    
    # train model
    roclist = []
    loss_list= []
    
    
    #torch.manual_seed(4)
    pbar = tqdm(range(1000),desc = "Training Epochs")
    for epoch in pbar:
        t = time.time()
    
        A_pred,Z1,Z2 = model(features1,features2)
        optimizer.zero_grad()
        loss = norm*F.binary_cross_entropy(A_pred.view(-1), adj_label.to_dense().view(-1), weight = weight_tensor)
    
        kl_divergence = 0.5/ A_pred.size(0) *( (1 + 2*model.logstd1 - model.mean1**2 - torch.exp(model.logstd1)**2).sum(1).mean()+
                                              (1 + 2*model.logstd2 - model.mean2**2 - torch.exp(model.logstd2)**2).sum(1).mean())
        loss -= kl_divergence
        loss.backward()
        optimizer.step()
        
    
        val_roc, val_ap = get_scores(val_edges, val_edges_false, A_pred)
        
        roclist.append(val_roc)
        loss_list.append(loss.item())
    
        pbar.set_postfix({"train_loss=": "{:.5f}".format(loss.item()),
                          'val_roc=': val_roc})
    
    
    A_pred,Z1,Z2 = model(features1,features2)
    
    AUC1, test_ap = get_scores(test_edges, test_edges_false, A_pred)
    print("1) End of training!", "test_roc=", "{:.5f}".format(AUC1),
          "test_ap=", "{:.5f}".format(test_ap))
    
    
    #torch.manual_seed(7)#16
    model2 = VBGAE(adj_norm,83,306)
    init_parameters(model2)
    #S = torch.Tensor(features01[:,Z]).reshape(-1,1)
    S1 = torch.Tensor(z1z).reshape(-1,1)
    S2 = torch.Tensor(z2z).reshape(-1,1)
    
    
    optimizer = Adam(model2.parameters(), lr=args.learning_rate)
    
    # train model
    roclist = []
    loss_list= []
    indep_list = []
    
    #torch.manual_seed(2)#3
    pbar = tqdm(range(1000),desc = "Training Epochs")
    for epoch in pbar:
        t = time.time()
    
        A_pred,Z1,Z2 = model2(features1,features2)
        optimizer.zero_grad()
        loss = norm*F.binary_cross_entropy(A_pred.view(-1), adj_label.to_dense().view(-1), weight = weight_tensor)
    
        kl_divergence = 0.5/ A_pred.size(0) *( (1 + 2*model2.logstd1 - model2.mean1**2 - torch.exp(model2.logstd1)**2).sum(1).mean()+
                                              (1 + 2*model2.logstd2 - model2.mean2**2 - torch.exp(model2.logstd2)**2).sum(1).mean())
        loss -= kl_divergence
        independance =2*(RFF_HSIC(model2.mean1,S1)) #+(RFF_HSIC(model2.mean2,S2)) #0.02
        #independance =2*torch.log(RFF_HSIC(Z1,S))#0.02
    
        loss += 1*independance
        loss.backward()
        optimizer.step()
        
    
        val_roc, val_ap = get_scores(val_edges, val_edges_false, A_pred)
        
        roclist.append(val_roc)
        loss_list.append(loss.item())
        indep_list.append(independance.item())
    
        pbar.set_postfix({"train_loss=": "{:.5f}".format(loss.item()),
                          'val_roc=': val_roc,'log_HSIC =': independance.item()})
    
    
    AUC2, test_ap = get_scores(test_edges, test_edges_false, A_pred)
    print("1) End of training!", "test_roc=", "{:.5f}".format(AUC2),
          "test_ap=", "{:.5f}".format(test_ap))
    
    
    stat1 = HSIC_stat(model2.mean1,S1)
    x = np.linspace(0, 0.4, 100)
    y = stats.gamma.cdf(x,stat1[3].item(),scale=stat1[4].item())
    p1=stats.gamma.sf(stat1[0].item()*n, stat1[3].item(), scale=stat1[4].item())
    
    stat1 = HSIC_stat(model2.mean2,S2)
    p2=stats.gamma.sf(stat1[0].item()*n, stat1[3].item(), scale=stat1[4].item())
    
    A_pred,Z1,Z2 = model(features1,features2)
    A_pred_fair,Z1,Z2 = model2(features1,features2)  
    
    
    SP01=scipy.stats.spearmanr(score0.detach().numpy().reshape(-1,1),A_pred.detach().numpy().reshape(-1,1))
    SP02=scipy.stats.spearmanr(score0.detach().numpy().reshape(-1,1),A_pred_fair.detach().numpy().reshape(-1,1))
    
    A_pred3 = distance_decode(model.mean1,model.mean2).detach().numpy()
    A_pred_fair3 = distance_decode(model2.mean1,model2.mean2).detach().numpy()
    
    SP1 = scipy.stats.spearmanr(score0.detach().numpy().reshape(-1,1),A_pred3.reshape(-1,1))
    SP2 = scipy.stats.spearmanr(score0.detach().numpy().reshape(-1,1),A_pred_fair3.reshape(-1,1))
    
    RES.iloc[k,] = [AUC1,AUC2,p1,p2,SP01[0],SP02[0], SP1[0],SP2[0]]
    print(RES.iloc[k,])
    print(RES.mean(0))
    print(1*(RES["SP1"][:k+1]<RES["SP2"][:k+1]).mean(0))

```
